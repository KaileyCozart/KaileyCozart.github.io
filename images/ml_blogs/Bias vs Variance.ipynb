```python
# Generated with Help from ChatGPT
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-3, 3, 100)

# Calculate y values for e^x and e^-x
y_exp = np.exp(x)
y_exp_neg = np.exp(-x)

# Calculate y values for the parabola
y_parabola = 2 * x**2 + 5

# Plotting the curves
plt.figure(figsize=(10, 6))
plt.plot(x, y_exp, label='Variance', color='blue')
plt.plot(x, y_exp_neg, label='Bias Squared', color='red')
plt.plot(x, y_parabola, label='Total Error', color='green')

# Adding labels and title
plt.xticks([])
plt.yticks([])
plt.xlabel('Model Complexity')
plt.ylabel('Error')
plt.title('Bias and Variance')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
